```python
#importing required libraries and dataset
import pandas as pd
from google.colab import drive
drive.mount('/content/drive', force_remount = True)
path = "drive/My Drive/Colab Notebooks/rainfall_by_districts_2020(1).csv"
df = pd.read_csv(path,index_col=0)
df.drop(["Normal Rainfall in South West Monsoon (June'2018 to September'2018) in mm","Normal Rainfall in North East Monsoon (October'2018 to December'2018) in mm","Normal Rainfall in Winter Season (January'2019 to February'2019) in mm","Normal Rainfall in Hot Weather Season (March'2019 to May'2019) in mm","Total Normal Rainfall (June'2018 to May'2019) in mm"],axis=1,inplace=True)
```

```python
#displaying dataset
df
```

```python
#dropping irrelevant data
df.drop('Total',inplace=True)
```

```python
#converting string values of district names to numerical data using labelencoder
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['District'] = le.fit_transform(df['District'])
df['District']
df
```

```python
new_df = df.drop(['District'],axis=1)
```

```python
df
```

```python
x = df[['District',"Total Actual Rainfall (June'2018 to May'2019) in mm"]].values
```

```python
y = df[["Actual Rainfall in South West Monsoon (June'2018 to September'2018) in mm",
       "Actual Rainfall in North East Monsoon (October'2018 to December'2018) in mm",
       "Actual Rainfall in Winter Season (January'2019 to February'2019) in mm",
       "Actual Rainfall in Hot Weather Season (March'2019 to May'2019) in mm"]]
```

```python
#importing libraries required to make the model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
model = LinearRegression().fit(x_train,y_train)
model.fit(x_train,y_train)
x_train
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
#accuracy of the model
import numpy as np
y_pred = pd.DataFrame(y_pred)
rmse = mean_squared_error(y_test, y_pred,squared=False)
print("Accuracy of the model is: ",rmse)
```

```python
label_district = df['District'].values
label_district
```

```python
b_df = pd.read_csv(path,index_col=0)
district = b_df['District'].values
total = b_df["Total Actual Rainfall (June'2018 to May'2019) in mm"].values
district
```

```python
dicts = {}
for i in range(len(district)-1):
  dicts[district[i]] = [label_district[i],total[i]]

print(dicts)
```

```python
#function to accept the input of the user of the district and printing the rainfall pattern
import matplotlib.pyplot as plt
def user():
    district= input("Enter the district name ").capitalize()
    climates = ["South west monsoon", "North east monsoon", "Winter season", "Hot winter season"]
    
    try:
      rainfall =model.predict([dicts[district]])[0]
      print(rainfall)
      print("Actual Rainfall in South West Monsoon (June'2018 to September'2018) in mm",round(rainfall[0],2))
      print("Actual Rainfall in North East Monsoon (October'2018 to December'2018) in mm",round(rainfall[1],2))
      print("Actual Rainfall in Winter Season (January'2019 to February'2019) in mm",round(rainfall[2],2))
      print("Actual Rainfall in Hot Weather Season (March'2019 to May'2019) in mm",round(rainfall[3],2))
      plt.figure(figsize = (10,5))
      plt.title("Rainfall by seasons")
      plt.xlabel("Seasons")
      plt.ylabel("Rainfall")
      plt.plot(climates,rainfall, marker = 'o', color = 'red')
    except:
      print("Please enter the correct input")
```
